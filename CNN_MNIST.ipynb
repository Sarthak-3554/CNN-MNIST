```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_final_test = x_test[-1]
y_final_test = y_test[-1]
```

```python
x_final_test
```

```python
x_test = x_test[:-1]
y_test = y_test[:-1]
```

```python
x_train[0]
```

```python
x_train = x_train/255

x_test = x_test/255
```

```python
x_train[0].shape
y_train[0:10]
```

```python
import keras
from keras import layers
```

```python
model = keras.Sequential()
```

```python
model.add(layers.Conv2D(filters=64,kernel_size=3,strides=1,padding="valid",activation="relu",input_shape=(28,28,1)))
```

```python
model.add( layers.MaxPool2D((2,2)))
```

```python
model.add( layers.Conv2D(128,3,1,padding="valid"))
```

```python
model.add( layers.MaxPool2D((2,2)))
```

```python
model.add(layers.Flatten())
```

```python
model.add(layers.Dense(3200,activation="sigmoid"))
```

```python
model.add(layers.Dense(10,activation="softmax"))
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.summary()
```

```python
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=32,
    validation_data=(x_test, y_test)
)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test accuracy:", test_acc*100)
```

```python
x_final_test = x_final_test/255
pred = model.predict(x_final_test.reshape(1,28,28,1))
pred.argmax()
```
